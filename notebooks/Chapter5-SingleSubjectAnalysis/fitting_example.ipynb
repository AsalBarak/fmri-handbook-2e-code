```python
from scipy.optimize import curve_fit
import numpy
import matplotlib.pyplot as plt
%matplotlib inline

def one_over_f(i,a):
    return 1./(i**a)

f=numpy.arange(0.01,1,0.01)
power_ts=one_over_f(f,0.5)
power_ts=power_ts+numpy.random.randn(len(power_ts))

fit=curve_fit(one_over_f,f,power_ts)
print fit
plt.plot(f,one_over_f(f,fit[0]))
plt.plot(f,power_ts)
```
